## Importação da base de dados via API

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json'

data = pd.read_json(url)
data.head()
```

### Normalização dos dados

```python
import requests
```

```python
r = requests.get(url)
data_json = r.json()
```

```python
data = pd.json_normalize(data_json)
```

## Tratamento de dados

```python
# Renomeação das colunas trocando '.' por '_'
data.columns = [column.replace('.', '_') for column in data.columns]
```

```python
# Visualização das colunas com Churn em branco
data[data['Churn'] == ''].head()
```

```python
# Exclusão das colunas com Churn em branco
data = data[data['Churn'] != '']
data.reset_index(drop=True, inplace=True)
```

```python
# Conferência das alterações
data['Churn'].value_counts()
```

```python
# Visualização das células com account_Charges_Total em branco
data[data['account_Charges_Total'] == ' '].head()
```

```python
# Visualização das células com customer_tenure com valor 0
data.query('customer_tenure == 0')[['account_Contract', 'customer_tenure', 'account_Charges_Monthly', 'account_Charges_Total' ]]
```

```python
# Index dos valores com account_Contract	== "Two year"
idx_two_year = data.query('customer_tenure == 0 and account_Contract	== "Two year"')[['account_Contract', 'customer_tenure' , 'account_Charges_Total', 'account_Charges_Monthly', ]].index
print(idx_two_year)
```

```python
# Index dos valores com account_Contract	== "One year"
idx_one_year = data.query('customer_tenure == 0 and account_Contract	== "One year"')[['account_Contract', 'customer_tenure' , 'account_Charges_Total', 'account_Charges_Monthly', ]].index
print(idx_one_year)
```

```python
# Inserção da quantidade de meses na coluna customer_tenure com account_Contract	== "Two year"
data.loc[idx_two_year, 'customer_tenure'] = 24
```

```python
# Inserção da quantidade de meses na coluna customer_tenure com account_Contract	== "One year"
data.loc[idx_one_year, 'customer_tenure'] = 12
```

```python
# Index dos valores com account_Charges_Total em branco
idx = data.query('account_Charges_Total == " "')[['account_Contract', 'customer_tenure' , 'account_Charges_Total', 'account_Charges_Monthly', ]].index
print(idx)
```

```python
# Inserção do valor total na coluna account_Charges_Total
data.loc[idx, 'account_Charges_Total'] = data.loc[idx, 'account_Charges_Monthly'] * data.loc[idx, 'customer_tenure']
```

```python
import numpy as np
```

```python
# Troca do tipo de dados da coluna account_Charges_Total para float
data['account_Charges_Total'] = data['account_Charges_Total'].astype(np.float64)
```

```python
# Criação da coluna com gastos diários
data['account_Charges_Daily'] = data['account_Charges_Monthly'] / 30
```
